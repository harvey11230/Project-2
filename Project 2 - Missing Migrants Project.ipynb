```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
mm = pd.read_csv('https://raw.githubusercontent.com/harvey11230/Project-2/master/MissingMigrantsProject.csv')
```

```python
mm = mm.dropna(how='any')
```

# Missing or dead migrants from 2014 to 2017 

```python
mm
```

```python
mm = mm[['incident_region','missing','dead']].copy()
```

```python
mm
```

# Sum the dead and missing imigrants in all incident regions

```python
groups = mm.groupby('incident_region').sum()
```

```python
groups.plot(kind='bar')
plt.xlabel('Missing or dead migrants from 2014 to 2017')
```

# The proportion of missing and dead immigrants in the Mediterranean is higher than in other regions.
