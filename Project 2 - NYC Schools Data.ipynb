```python
import pandas as pd
```

```python
schools = pd.read_csv('https://raw.githubusercontent.com/harvey11230/Project-2/master/NYC%20Schools%20Data.csv')
```

```python
schools = schools.dropna(how='any')
```

# read data and dropped any records contains NAN

```python
schools
```

```python
schools = schools[['school_name','SAT Critical Reading Avg. Score','SAT Math Avg. Score','SAT Writing Avg. Score']].copy()
```

# After read data, I recongized there was no neeed to do calculation. I plan to visualize the top 5 high school's SAT average scores on plot.

```python
top5 = schools.sort_values(["SAT Critical Reading Avg. Score", "SAT Math Avg. Score", "SAT Writing Avg. Score"], ascending= [False, False, False]).head()
```

```python
top5.set_index('school_name', inplace=True)
```

# I use school name instead index that could help me easier to build dictionary.

```python
top5.to_dict()
```

```python
import matplotlib.pyplot as plt
```
